# Wikipedia List

```
import requests
page = requests.get("https://en.wikipedia.org/wiki/List_of_video_games_considered_the_best")
page
```

```
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
table = soup.find(class_='sortable')

ol = soup.findAll('ol')
lis = ol[1].findAll('li')

master = list()
for li in lis:
  master.append((
      li.get('id'), li.text.strip().replace('^ ', '').count('[')
  ))

m = dict(master)

rows = list()
current_year = 0

for items in table.findAll('tr'):
  year = items.find('th')
  if (year != None):
    current_year = year.text.strip()

  item = items.findAll('td')
  if (len(item) == 5):
    rows.append({
        'year': current_year,
        'title': item[0].text.strip(),
        'genre': item[1].text.strip(),
        'publisher': item[2].text.strip(),
        'plataform': item[3].text.strip(),
        'refs': m[item[4].find(href=True)['href'].replace('#', '')]
    })

import pandas as pd
df = pd.DataFrame(rows)

df.to_csv('video_games_considered_the_best.csv') 

print(df.count())
pd.set_option('max_rows', 99999)
df.sort_values('refs', ascending=False)
```

```
publishers = df.groupby('publisher').size()

publishers_sorted = sorted(dict(publishers).items(), key=lambda x: x[1], reverse=True)
df = pd.DataFrame(publishers_sorted)
df.columns = [ 'publisher', 'games']
df
```

```
teste = df.groupby('genre').size()

xx = sorted(dict(teste).items(), key=lambda x: x[1], reverse=True)
x = pd.DataFrame(xx)
x.columns = [ 'genre', 'games']
x
```

```
teste = df.groupby('plataform').size()

xx = sorted(dict(teste).items(), key=lambda x: x[1], reverse=True)
x = pd.DataFrame(xx)
x.columns = [ 'plataform', 'games']
x
```

```
stealth_games = df[df['genre'] == 'Stealth']
stealth_games
```
