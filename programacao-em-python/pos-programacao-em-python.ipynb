```
# Exercício de dicionário
pessoas = []
quantidade_de_pessoas = 3
mulheres = []

while len(pessoas) < quantidade_de_pessoas:
  print('-'*30)
  pessoas.append({
      'nome': str(input('Digite o nome: \n')),
      'sexo': str(input('Digite o sexo: [F/M]\n'))[0].upper(),
      'idade': int(input('Digite a idade: \n'))
  })

print('-'*30)
print('Todas as pessoas:')
for pessoa in pessoas:
  print('Nome: ' + pessoa['nome'] + ', sexo: ' + pessoa['sexo'] + ', idade: ' + str(pessoa['idade']))

print('-'*30)
print('Apenas as mulheres:')
for pessoa in pessoas:
  if pessoa['sexo'] == 'F':
    print('Nome: ' + pessoa['nome'] + ', sexo: ' + pessoa['sexo'] + ', idade: ' + str(pessoa['idade']))

print('-'*30)
media_das_idades = sum(pessoa['idade'] for pessoa in pessoas) / quantidade_de_pessoas
print(f'A média das idades é: {media_das_idades:.2f}')

print('-'*30)
pessoas_com_idade_acima_da_media = []
for pessoa in pessoas:
  if pessoa['idade'] > media_das_idades:
    pessoas_com_idade_acima_da_media.append(pessoa['nome'])

if len(pessoas_com_idade_acima_da_media) > 0:
  print('Pessoas com idade acima da média: ')
  print(pessoas_com_idade_acima_da_media)
else:
  print('Não existem pessoas com idade acima da média')
```

```
# Exercício de lista 2
lista_geral = []
lista_temporaria_de_pares = []
lista_temporaria_de_impares = []

while True:
  numero = int(input('Digite um número: \n'))
  if numero % 2 == 0:
    lista_temporaria_de_pares.append(numero)
  else:
    lista_temporaria_de_impares.append(numero)
  
  op = str(input('Deseja continuar? [S/N]'))[0].upper()
  if op == 'N':
    break


lista_temporaria_de_impares.sort()
lista_temporaria_de_pares.sort(reverse=True)

lista_geral.append(lista_temporaria_de_pares.copy())
lista_geral.append(lista_temporaria_de_impares.copy())

print('Lista geral:')
print(lista_geral)
print(f'Primeiro valor da lista par: {lista_geral[0][0]}')
print(f'Último valor da lista ímpar: {lista_geral[-1][-1]}')
```

```
# Função aplicada em python
def cria_lista():
  lista_de_numeros = []

  while True:
    numero = int(input('Digite um número: \n'))
    lista_de_numeros.append(numero)

    op = str(input('Deseja continuar? [S/N]\n'))[0].upper()
    if op == 'N':
      break
  
  return lista_de_numeros

def verifica_se_lista_esta_ordenada(lista):
  if lista == sorted(lista):
    return 'Lista ordenada'
  else:
    return 'Lista não ordenada'

def ordena_e_exibe_a_lista(lista):
  if verifica_se_lista_esta_ordenada(lista) == 'Lista não ordenada':
    lista.sort() 
  
  print(lista)


lista = cria_lista()
ordena_e_exibe_a_lista(lista)
```

```
import numpy as np

numeros = []
quantidade_de_linhas = int(input('Digite a quantidade de linhas que a matriz vai ter:\n'))
quantidade_de_colunas = int(input('Digite a quantidade de colunas que a matriz vai ter:\n'))
quantidade_de_numeros_da_matriz = quantidade_de_linhas * quantidade_de_colunas
print('-' * 50)
print(f'Digite {quantidade_de_numeros_da_matriz} números para formar a matriz')

for numero in range(quantidade_de_numeros_da_matriz):
  numeros.append(int(input(f'Digite o número {numero}:\n')))

matriz = np.array(numeros).reshape(quantidade_de_linhas, quantidade_de_colunas)
print(matriz)

while True:
  print('-' * 50)
  print('Digite os valores para fatiar a matriz: \n')
  linha_inicial = int(input(f'Digite a linha inicial entre 0 e {quantidade_de_linhas}: \n'))
  if linha_inicial < 0 or linha_inicial > quantidade_de_linhas:
    print('Valor da linha inicial inválido! Tente novamente')
    continue
  
  ultima_linha = int(input(f'Digite a última linha, entre {linha_inicial} e {quantidade_de_linhas}: \n'))
  if ultima_linha < linha_inicial or ultima_linha > quantidade_de_linhas:
    print('Valor da última linha inválido! Tente novamente')
    continue

  coluna_inicio = int(input(f'Digite a coluna inicial entre 0 e {quantidade_de_colunas}: \n'))
  if coluna_inicio < 0 or coluna_inicio > quantidade_de_colunas:
    print('Valor da coluna de inicio inválido! Tente novamente')
    continue

  coluna_fim = int(input(f'Digite última coluna entre {coluna_inicio} e {quantidade_de_colunas}: \n'))
  if coluna_fim < coluna_inicio or coluna_fim > quantidade_de_colunas:
    print('Valor da última coluna inválido! Tente novamente')
    continue

  print('Matriz fatiada: \n')
  print(matriz[linha_inicial:ultima_linha, coluna_inicio:coluna_fim])
  break
```
