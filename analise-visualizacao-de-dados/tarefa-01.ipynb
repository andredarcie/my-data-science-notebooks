## Análise das variáveis

- **date**: Date de coleta dos dados no formato YYYY-MM-DD.
- **state**: Sigla da unidade federativa.
- **city**: Nome do município.
- **place_type**: Tipo de local que esse registro descreve, pode ser city ou state.
- **confirmed**: Número de casos confirmados.
- **deaths**: Número de mortes.                          
- **order_for_place**: Número que identifica a ordem do registro para este local. O registro referente ao primeiro boletim em que esse local aparecer será contabilizado como 1 e os demais boletins incrementarão esse valor.
- **is_last**: Campo pré-computado que diz se esse registro é o mais novo para esse local.                     
- **estimated_population_2019**: População estimada para esse município/estado em 2019.
- **city_ibge_code**: Código IBGE do local.
- **confirmed_per_100k_inhabitants**: Número de casos confirmados por 100.000 habitantes.
- **death_rate**: Taxa de mortalidade (mortes / confirmados).

```
print('Casos confirmados por estado: \n')
df[(df['is_last'] == True) & (df['place_type'] == 'state')][['state', 'confirmed']]
```

```
print('Estado com maior número de casos confirmados:')
df[df['confirmed'] == df['confirmed'].max()][['state', 'confirmed']]
```

```
print('Mortes por estado: \n')
df[(df['is_last'] == True) & (df['place_type'] == 'state')][['state', 'deaths']]
```

```
print('Estado com maior número de mortes:')
df[df['confirmed'] == df['confirmed'].max()][['state', 'deaths']]
```

```
print('Casos confirmados sem morte: ')
df['casos sem morte'] = df['confirmed'] - df['deaths']
df.sample(5)
```

```
print('Casos confirmados e mortes por cidades')
df[(df['is_last'] == True) & (df['place_type'] == 'city')][['city', 'confirmed', 'deaths']]
```

```
cidades = df.loc[df.city.notnull()]
cidades = cidades.loc[cidades.is_last == True]
cidades.sort_values(by=['confirmed'], ascending=False)
```

```
print('Total de casos confirmados: ')
cidades.confirmed.sum()
```

```
print('Total de óbitos: ')
cidades.deaths.sum()
```
