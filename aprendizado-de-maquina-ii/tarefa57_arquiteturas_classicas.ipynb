# Importa as dependencias necessarias

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing import image
```

## Faz upload do arquivo para o google colab

```python
from google.colab import files
uploaded = files.upload()
filename = next(iter(uploaded))
```

# Imagem de um pão de queijo

```python
from IPython.display import Image
filepath = '/content/'+filename
Image(filepath)
```

```python
from tensorflow.keras.applications.inception_resnet_v2 import InceptionResNetV2, preprocess_input, decode_predictions
```

## Testando a predição na rede InceptionResNetV2

```python
model = InceptionResNetV2(weights='imagenet')
img_path = filename
img = image.load_img(img_path, target_size=(299, 299))
image_array = image.img_to_array(img)
image_array = np.expand_dims(image_array, axis=0)
image_array = preprocess_input(image_array)

preds = model.predict(image_array)
print('Resultado: ', decode_predictions(preds, top=3)[0])
```

## Treinando a predição na rede DenseNet201

```python
from tensorflow.keras.applications.densenet import DenseNet201, preprocess_input, decode_predictions
```

```python
model = DenseNet201(weights='imagenet')
img_path = filename
img = image.load_img(img_path, target_size=(224, 224))
image_array = image.img_to_array(img)
image_array = np.expand_dims(image_array, axis=0)
image_array = preprocess_input(image_array)

preds = model.predict(image_array)
print('Resultado: ', decode_predictions(preds, top=3)[0])
```
